```python
import pylsl

streams = pylsl.resolve_streams()
for stream in streams:
    print(stream.name())

# get chunks of data from the first stream
inlet = pylsl.StreamInlet(streams[0])
while True:
    chunk, timestamps = inlet.pull_sample()
    if timestamps:
        print(timestamps, chunk)
```
